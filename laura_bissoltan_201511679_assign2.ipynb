```python
#1.1

from math import factorial,e 

#first_approach
def exps_one(x, n):
    term_approx = 0
    for i in range(n):
        term_approx += ((-1)**i) * ((x**i)/(factorial(i)))
    return term_approx

#second_approach
def exps_two(x, n):
    term_approx = 0
    for i in range(n):
        term_approx += ((x**i)/(factorial(i)))
        val = 1/term_approx
    return val

print("Table 1: Evaluation of e^x at x=-5 using first method")
print(' {:4} {:15} {:15} \033[0m'.format('i','   \u03B5_a(%)',' \u03B5_t(%)'))

t_v = e**-(5)
i = 2
while i <= 20:
    eps_a = (abs((exps_one(5,i)-exps_one(5,i-1))/exps_one(5,i))*100)
    eps_t = abs((t_v - exps_one(5,i))/t_v)*100
    print(f'{i:2}  {eps_a:12.6f} {eps_t:16.6f}')
    i = i+1
    

print("\nTable 1: The energies of the first six energy levels of the particle")
print(' {:4} {:15} {:15} \033[0m'.format('i',' \u03B5_a(%)','  \u03B5_t(%)'))

i = 2
while i <= 20:
    eps_a = (abs((exps_two(5,i)-exps_two(5,i-1))/exps_two(5,i))*100)
    eps_t = abs((t_v - exps_two(5,i))/t_v)*100
    print(f'{i:2} {eps_a:12.6f} {eps_t:16.6f}')
    i = i+1
    

```

```python
#2.1

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 0.25 
x = np.linspace(-2,2)

def f(x):
    return 5*x**3-2*x+4

def dfdx(x):
    return 15*x**2-2

#forward difference of f'(x) and f''(x)
def dfdhf1(x,h):
    return (f(x+h)-f(x))/h
 
#backward difference of f'(x) and f''(x)
def dfdhb1(x,h):
    return (f(x)-f(x-h))/h    

#central difference of f'(x) and f''(x)
def dfdhc1(x,h):
    return (f(x+h)-f(x-h))/(2*h)    

yp = dfdx(x)
yp1 = dfdhf1(x,h)
yp2 = dfdhb1(x,h)
yp3 = dfdhc1(x,h)

plt.plot(x,yp,'m--',x,yp1,'g--',x,yp2,'c--',x,yp3,'r+')
plt.xlabel(r'$x$',size=18)
plt.ylabel(r'$f(x)$',size=18)
plt.legend(["f'(x)", "ff'(x)", "fb'(x)", "fc'(x)"])
plt.grid()
plt.show()
    
```

```python
#2.2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 0.25 
x = np.linspace(-2,2)

def f(x):
    return 5*x**3-2*x+4

def dfdx2(x):
    return 30*x

#forward difference of f'(x) and f''(x)
def dfdhf2(x,h):
    return (f(x+2*h)-2*f(x+h)+f(x))/h**2
 
#backward difference of f'(x) and f''(x)
def dfdhb2(x,h):
    return (f(x)-2*f(x-h)+f(x-2*h))/h**2

#central difference of f'(x) and f''(x)
def dfdhc2(x,h):
    return (f(x+h)-2*f(x)+f(x-h))/h**2

yv = dfdx2(x)
yv1 = dfdhf2(x,h)
yv2 = dfdhb2(x,h)
yv3 = dfdhc2(x,h)

plt.plot(x, yv,'m--', x, yv1,'g--', x, yv2,'c--',x, yv3,'r+')
plt.xlabel(r'$x$', size=18)
plt.ylabel(r'$f(x)$', size=18)
plt.legend(["f''(x)", "ff''(x)", "fb''(x)", "fc''(x)"])
plt.grid()
plt.show()
```

```python
#3.2

%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib inline
from matplotlib import cm 
import numpy as np

a1=15.67
a2=17.23
a3=0.75
a4=93.5
 
#binding_energy_per_nucleon

def binding_energy_pn(N, Z):    
   
    A=Z+N
    
    if (A%2==1).any():
        a5=0
    elif (Z%2==0 and A%2==0).any():
        a5=12
    elif (A%2==0 and Z%2==1).any():
        a5=-12 
        
    E_B = a1*A-a2*A**(2/3)-a3*Z**2/A**(1/3)-a4*(A-2*Z)**2/A+a5/A**(1/2)
    return E_B/A
   

fig = plt.figure(dpi=300)
ax = fig.gca(projection='3d')

N = np.linspace(0,150,30)
Z = np.linspace(1,100,30)
NN,ZZ = np.meshgrid(N, Z)
B = binding_energy_pn(NN,ZZ)

surf = ax.plot_surface(ZZ, NN, B, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_xlabel('$Z$', size=14)
ax.set_ylabel('$N$', size=14)
ax.set_zlabel(r'$E_B/A (MeV)$', size=14)

phi=10 
theta=60
ax.view_init(phi, theta)


fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_title("The binding energy per nucleon(E_B/A) vs the atomic " \
"number(Z) vs the atomic number(N)")

from matplotlib.ticker import LinearLocator, FormatStrFormatter
# Customize the z axis.
ax.set_zlim(-100, 60)
ax.zaxis.set_major_locator(LinearLocator(6))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.01f'))
plt.show()


```
